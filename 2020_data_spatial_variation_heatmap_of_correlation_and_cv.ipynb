```python
import numpy as np
import scipy.stats as st
import pandas as pd
```

```python
#data = pd.read_csv('firetowerFiber_USDA.csv')
#data = pd.read_csv('FT_join_file.csv')
data = pd.read_excel('bighalf_all_data.xlsx')
```

```python
data
```

```python
fiber_data = data.iloc[:, 23:]
fiber_data.describe()
```

```python
#define function to calculate cv
cv = lambda x: np.std(x, ddof=1) / np.mean(x) * 100 
#calculate CV for each column in data frame
fiber_data.apply(cv)
```

```python
import seaborn as sns
corr_bf = fiber_data.corr()


fig, ax = plt.subplots(figsize=(10, 8))
# mask
mask = np.triu(np.ones_like(corr_bf, dtype=np.bool))

# adjust mask and df
mask = mask[1:, :-1]
corr_bf = corr_bf.iloc[1:,:-1].copy()

# color map
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
# plot heatmap
sns.heatmap(corr_bf, mask=mask, annot=True, fmt=".2f", 
           linewidths=5, cmap=cmap, vmin=-1, vmax=1, 
           cbar_kws={"shrink": .8}, square=True)
# ticks
yticks = [i for i in corr_bf.index]
xticks = [i for i in corr_bf.columns]
plt.yticks(plt.yticks()[0], labels=yticks, rotation=0)
plt.xticks(plt.xticks()[0], labels=xticks)
# title
title = 'CORRELATION MATRIX at big half field in 2020\n'
plt.title(title, loc='left', fontsize=18)
plt.show()
```
